```python
import os
import re
import shutil
source_directory_path = 'C:/Users/Lenovo/PycharmProjects/HCI_Project/class/3-class'
destination_directory_path = 'C:/Users/Lenovo/PycharmProjects/HCI_Project/Data'

prefix_mapping = {
    'yukari': 'Up',
    'asagi': 'Down',
    'sag': 'Right',
    'sol': 'Left',
    'kirp': 'Blink'
}

if not os.path.exists(destination_directory_path):
    os.makedirs(destination_directory_path)
    print(f"Created destination directory '{destination_directory_path}'")

others_directory_path = os.path.join(destination_directory_path, "others")
if not os.path.exists(others_directory_path):
    os.makedirs(others_directory_path)
    print(f"Created 'others' subdirectory at '{others_directory_path}'")


def count_files(directory):
    return len([name for name in os.listdir(directory) if os.path.isfile(os.path.join(directory, name))])


initial_source_count = count_files(source_directory_path)
initial_destination_count = count_files(destination_directory_path)
initial_others_count = count_files(others_directory_path)
print(
    f"Initial count - Source: {initial_source_count}, Destination: {initial_destination_count}, Others: {initial_others_count}")

files = os.listdir(source_directory_path)


def map_filename(old_name):
    for key, value in prefix_mapping.items():
        if re.match(f'^{key}', old_name, re.IGNORECASE):
            return re.sub(f'^{key}', value, old_name, flags=re.IGNORECASE)
    return None


others_count = 0
for file in files:
    old_file_path = os.path.join(source_directory_path, file)
    if os.path.isfile(old_file_path):
        new_name = map_filename(file)
        if new_name:
            new_file_path = os.path.join(destination_directory_path, new_name)
            if os.path.exists(new_file_path):
                base, ext = os.path.splitext(new_file_path)
                new_file_path = f"{base}_2{ext}"
        else:
            new_file_path = os.path.join(others_directory_path, file)
            others_count += 1
        shutil.move(old_file_path, new_file_path)
        print(f"Moved '{file}' to '{new_file_path}'")
final_source_count = count_files(source_directory_path)
final_destination_count = count_files(destination_directory_path)
final_others_count = count_files(others_directory_path)
print(
    f"Final count - Source: {final_source_count}, Destination: {final_destination_count}, Others: {final_others_count}")
print(f"Total files moved to 'others': {others_count}")

print("Operation complete.")
```
